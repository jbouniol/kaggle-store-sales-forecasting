```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


from prophet import Prophet
```

```python
train = pd.read_feather('../Data/Processed/train.feather')
test = pd.read_feather('../Data/Processed/test.feather')
```

```python
pd.set_option('display.max_columns', None)
train.head()
```

```python
train_prophet = train.rename(columns={'date': 'ds', 'sales': 'y'})

X_train, X_test, y_train, y_test = train_test_split(train_prophet[['ds']], train_prophet['y'], test_size=0.2, random_state=42)

model = Prophet()

train_data = pd.DataFrame({'ds': X_train['ds'], 'y': y_train})

model.fit(train_data)

test_data = pd.DataFrame({'ds': X_test['ds']})

forecast = model.predict(test_data)
```
